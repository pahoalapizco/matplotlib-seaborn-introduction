# Set

Con esta función podemos definir un tema especifico, tipo de fuente, tamaño de fuente, etc. Es sumamente útil para personalizar los graficos que deseamos exponer.

**Importante tener en cuenta**
- Este parámetro afecta a todas las gráficas que creamos, incluidas las que no usan Seaborn.
- Podemos resetear los valores utilizando el método reset_orig.
- Este parámetro es un atajo de set_theme, que es el recomendado según la documentación

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Cambiar fuentes
from matplotlib.font_manager import fontManager,FontProperties
```

Registrar la fuente

```python
path = "../fonts/verdana.ttf"
fontManager.addfont(path)

prop=FontProperties(fname=path)
sns.set(font=prop.get_name())
```

**Grafica default**

```python
sns.barplot(x=['A', 'B', 'C'], y=[1, 3, 2])
plt.show()
```

## Tema

| Style |
|-------|
|darkgrid|
|whitegrid|
|dark|
|white|
|ticks|

```python
sns.set(style='whitegrid')
sns.barplot(x=['A', 'B', 'C'], y=[1, 3, 2])
plt.show()
```

## Paleta de colores

Con el parametro `palette` podemos cambiar la tonalidad de las graficas
|Palette|
|-------|
|deep|
| muted|
| bright|
| pastel|
| dark|
| colorblind|

```python
sns.set(style='darkgrid', palette='colorblind', )
sns.barplot(x=['A', 'B', 'C'], y=[1, 3, 2])
plt.show()
```

## Fuente

### Tamaño

```python
sns.set(style='darkgrid', palette='bright', font_scale=5.5)
sns.barplot(x=['A', 'B', 'C'], y=[1, 3, 2])
plt.show()
```

### Tipo de fuente

```python
sns.set(style='darkgrid', palette='bright', font='Verdana', font_scale=2.5)
sns.barplot(x=['A', 'B', 'C'], y=[1, 3, 2])
plt.show()
```

Los cambios de estilo, colores y fuentes tienen sus propias funciones independientes que sustituyen a `set`, en proximas versiones esta función quedará descontinuada, así que a continuación dejo ejemplos de las funciones para estilo, color y fuente.

## set_theme()

```python
sns.set_theme(style='white', palette='pastel')
sns.barplot(x=['A', 'B', 'C'], y=[1, 3, 2])
plt.show()
```

# set_palette() y color_palette()

```python
sns.set_palette('dark')
sns.barplot(x=['A', 'B', 'C'], y=[1, 3, 2])
plt.show()
```

```python
sns.set_palette('Spectral')
sns.barplot(x=['A', 'B', 'C'], y=[1, 3, 2])
plt.show()
```

```python
sns.color_palette("Spectral", 19) # Paleta de colores, el 2do paraetro es el número de colores de la paleta que queremos ver
```

# set_style()

```python
sns.set_style("whitegrid", {"grid.color": ".6", "grid.linestyle": ":"})
sns.barplot(x=["A", "B", "C"], y=[1, 3, 2])
```
