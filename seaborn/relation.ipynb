# Relación

Estos tipos de gráficos son útiles para explorar la relación entre dos variables numéricas, nos permite visualizar patrones, tendencias y correlación de los datos.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

## Scatter

Este gráfico traza los valores de dos variables numéricas como coordenas en un plano(espacio) bidimensional, cada punto en la gráfica representa una observación.

Parámetros más comunes:
 - **`data`**: DataFrame con los datos.
 - **`x`**: Nombre de la columna del DataFrame de donde tomará el valor para la coordenada `x`.
 - **`y`**: Nombre de la columna del DataFrame de donde tomará el valor para la coordenada `y`.
 - **`hue`**: Nombre de la columna del DataFrame con los datos por los cuales queremos segmentar la gráfica.
 - **`style`**: Cambia el estilo de los puntos (cooredenadas) de la gráfica, utiliza los datos de la columna especificada, es otra forma de segmentar los datos graficados.
 - **`size`**: Cambia el tamaño de los puntos en la gráfica, utiliza los datos de la columna especificada.

Esté scatter no muestra la correlación entre las variables `total_bill` y `tip`:

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.show()
```

Esté scatter no muestra la correlación entre las variables `total_bill` y `tip` segmentadas por el día:

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day')
plt.show()
```

Utilizamos el `time` como estilo para cada punto del plano, de esta forma segmentamos la correlacion entre `total_bill` y `tips` por la hora en que se realizó el evento.

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time')
plt.show()
```

Al utilizar el parametro `size` cambiaremos el tamaño de los puntos en el plano, del dataset le mandamos el nombre de la columna `size` el cual contiene la cantidad de comensales por mesa:

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size')
plt.show()
```

Para evitar saturaración en el gráfico cambiaremos sus dimensiones y la posición del recuadro de leyendas.

```python
plt.figure(figsize=(6,6))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size')
plt.legend(loc='center', bbox_to_anchor=(1.12, 0.5))
plt.show()
```

## Line plot

Muestra la reación entre dos variables numéricas como una línea que conecta los puntos en una secuencia ordenada. Este tipo de gráfico es útil para visualizar la tendencia de una variable numérica en función de otra variable numérica.

Parámetros más comunes:
 - **`data`**: DataFrame con los datos.
 - **`x`**: Nombre de la columna del DataFrame de donde tomará el valor para la coordenada `x`.
 - **`y`**: Nombre de la columna del DataFrame de donde tomará el valor para la coordenada `y`.
 - **`hue`**: Nombre de la columna del DataFrame con los datos por los cuales queremos segmentar la gráfica.
 - **`style`**: Cambia el estilo de los puntos (cooredenadas) de la gráfica, utiliza los datos de la columna especificada, es otra forma de segmentar los datos graficados.
 - **`size`**: Cambia el tamaño de los puntos en la gráfica, utiliza los datos de la columna especificada.

Con el siguiente `lineplot` graficamos la relación entre las variabeles `total_bill` y `tip`:

```python
sns.lineplot(data=tips, x='total_bill', y='tip')
plt.show()
```

Al igual que con la función `scatterplot` podemos agregar segmentos, estilos y tamaños a las tiendas que representan la información.

```python
plt.figure(figsize=(6,6))
sns.lineplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size')
plt.legend(loc='center', bbox_to_anchor=(1.12, .5))
plt.show()
```

## Rel plot

Similar a las funciones `displot` y `catplot`, `replot` es de uso general y en ella podemos graficar N tipos de graficos según los parametros que le enviemos.

Parámetros más comunes:
 - **`data`**: DataFrame con los datos.
 - **`x`**: Nombre de la columna del DataFrame de donde tomará el valor para la coordenada `x`.
 - **`y`**: Nombre de la columna del DataFrame de donde tomará el valor para la coordenada `y`.
 - **`hue`**: Nombre de la columna del DataFrame con los datos por los cuales queremos segmentar la gráfica.
 - **`style`**: Cambia el estilo de los puntos (cooredenadas) de la gráfica, utiliza los datos de la columna especificada, es otra forma de segmentar los datos graficados.
 - **`size`**: Cambia el tamaño de los puntos en la gráfica, utiliza los datos de la columna especificada.
 - **`kind`**: Tipo de grafico que deseamos ilustrar.
 - **`col`**: Nombre de la columna por la cual queremos separar las grafias en graficos distintos.

Por default `relplot` muestra una grafica de scatter:

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size')
plt.show()
```

Con `col` generamos columnas respecto a los tiempos:

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', col='time')
plt.show()
```

O bien utilizamos `col` para generamos columnas respecto a si los comensales son fumadores o no:

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', col='smoker')
plt.show()
```
