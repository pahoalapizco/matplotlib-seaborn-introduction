# Subplots

`subplots` (plural) a diferencia de `subplot` (singular), aprobecha las bondades del método orientado a objetos.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = np.sin(x)
```

La función `subplots` regresa una figura y un array de axes del tamaño que se le indique en las filas y columnas. 

```python
figure, axes = plt.subplots(nrows=1, ncols=2)
```

Al tener los axes como arrays es posible acceder a ellos por medio de su indice.

```python
figure, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y, 'r')
axes[1].plot(y, x, 'b')
```

Otra forma de obtener los axes:

```python
figure, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x, y, 'r')
ax2.plot(y, x, 'b')
```

El resultado es el mismo, pero la forma de obtener los axes es distinto.

Al llamar a `subplots` podemos omitir el nombre de los parametros y simplemente enviar el total de filas y columnas.

```python
figure, axes = plt.subplots(2, 4)
```

```python
figure, axes = plt.subplots(2, 4)
axes[0, 0].plot(x, y, 'r')
axes[0, 1].plot(x, np.cos(x), 'b')
axes[0, 2].plot(x, np.tan(x), 'b')
axes[0, 3].plot(x, np.sin(x)**2, 'b')
```

```python
figure, (axes_r1, axes_r2) = plt.subplots(2, 4)
axes_r1[0].plot(x, y, 'r')
axes_r1[1].plot(x, np.sin(x))
axes_r1[2].plot(x, np.cos(y), 'm:')
axes_r1[3].plot(x, np.tan(y))

axes_r2[0].plot(np.cos(x), y, '1--')
axes_r2[1].plot(np.sin(y), x)
axes_r2[2].plot(np.tan(x), x, 'g')
axes_r2[3].plot(np.sin(x)**2, x, 'y:2')

figure.tight_layout() # crea un gap entre cada grafico
```
