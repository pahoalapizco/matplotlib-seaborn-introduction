# Pyplot básico

`pyplot` es una módulo de la libreria `matplotlib` con la cual podemos crear graficos simples en los ejes `x` y `y`. Las representaciones pueden ser líneas, imagenes, texto, etc.

```python
import matplotlib.pyplot as plt
import numpy as np
```

Con `numpy` creamos vectores con datos aleatorios para los ejex `x` y `y`

```python
x = np.linspace(0, 5, 11)
y = x ** 2
print(f'x: ', x)
print(f'y: ', y)
```

## plot(x_axis, y_axis)

La función `plot` dibuja un polígono con los los vértices dados las coordenadas `(x,y)` que se envian como parametros.
Donde `x` y `y` son vectores.

```python
plt.plot(x, y) # Crea el grafico
plt.show() # Muestra el grafico
```

Opcionalmente podemos formatear el grafico enviandole el parametro `[fmt]` (formating) que recibe un valor string, el cuál modifica color, marcador y línea de la grafica.
Posibles combinaciones:
1. '[marcador][línea][color]'
2. '[color][marcador][línea]'

Cada uno de los elemetos es opcional.

### Cambiar color

```python
plt.plot(x, y, 'g') # Crea el grafico con la línea color verde
plt.show() # Muestra el grafico
```

### Cambiar marcadores

```python
plt.plot(x, y, 'mo') # Crea el grafico con la línea color morada y un marcador de tipo circulo
plt.show() # Muestra el grafico
```

### Cambiar estilo línea

```python
plt.plot(x, y, 'y:x') # Crea el grafico con la línea color amarilla, un marcador de tipo cruz y la línea punteada
plt.show() # Muestra el grafico
```

## hist(x)

Dibuja un histograma con las frecuencias de los valores agrupados de `x` (vector/lista).

```python
plt.hist(x) # Crea el grafico
plt.show() # Muestra el grafico
```

## pie(x)

Dibuja un gráfico de pastel con las frecuencias de los valores en `x`.

```python
plt.pie(x) # Crea el grafico
plt.show() # Muestra el grafico
```

Con el parametro `radius` le indicamos el tamaño del radio del circulo, entre más grande el radio más grande será la gráfica.

```python
plt.pie(x, radius=3) # Crea el grafico
plt.show() # Muestra el grafico
```

## boxplot(x)

En español conocidida como `gráfica de caja y bigotes`, dibuja la grafica con los datos de la lista de `x`.

```python
plt.boxplot(x) # Crea el grafico
plt.show() # Muestra el grafico
```
