# Bar plot

En espa√±ol "grafica de barras", este tipo de barras es utilizado para mostrar datos categoricos (texto).

```python
import matplotlib.pyplot as plt
import numpy as np
```

Generamos datos de prueba en dos diferentes arrays. Uno de ellos de tipo string que iidentifica las categorias.

```python
countries = ['MEXICO', 'GERMANY', 'USA', 'CANADA', 'BRAZIL']
population = [10000, 9500, 11000, 11009, 13500]
```

### Grafica de barras default

```python
plt.bar(countries, population)
plt.show()
```

**Moficiando colores`(color)` y tama√±o `(width)`**
- `color`: String o array de strings
- `width`: Valor num√©rico con valor default en `0.8`.

```python
plt.bar(countries, population, color=['blue', 'purple'], width=0.5) # Alterna colores
plt.show()
```

**Modificar nombre de las categorias `(xticks)` y su orientaci√≥n `(rotation)`**
- `xticks`: Funci√≥n que recibe un array con las posiciones y otro con los nuevos t√≠tulos que llevara cada posici√≥n
- `rotation`: Par√°metro que recibe un valor num√©rico el c√∫l indica el angulo de rotaci√≥n del texto que se encuentra en el eje `X`, esta propiedad es de gran utilidad cuando se tienen textos muy largos o muchas barras.

```python
plt.bar(countries, population, color=['blue', 'purple'], width=0.5) # Alterna colores
plt.xticks(np.arange(len(countries)), ['Mexico', 'Alemania', 'EE.UU.', 'Canada', 'Brasil'], rotation=45) # rotaci√≥n de 45 grados
plt.show()
```

### Grafico de barras horizonal

```python
plt.barh(countries, population, color=['blue', 'purple'], height=0.5) # Alterna colores
plt.show()
```

El grosor de las barras horizontales se cambia con el par√°metro `height` (altura) por que el `width` esta dado por los valores correspondientes a las variables de texto, en el ejemplo anterior por el array `population`.

#### T√≠tulos, etiquetas y leyendas

```python
spanish =  ['Mexico', 'Alemania', 'EE.UU.', 'Canada', 'Brasil']
idxs = np.arange(len(countries))

plt.bar(countries, population, color=['blue', 'purple'], width=0.5) # Alterna colores
# Titulos
plt.title('Poblacion por pais')
plt.xlabel('Pais')
plt.ylabel('Poblacion')
plt.xticks(idxs, spanish, rotation=45) # rotaci√≥n de 45 grados
plt.show()
```

#### Grafica de barras con subplots

```python
fig, (ax1, ax2) = plt.subplots(2)
ax1.set_title('Grafica vertical')
ax1.bar(countries, population, color='#DD11DD')

ax2.set_title('Grafica vertical')
ax2.barh(countries, population)

fig.tight_layout() 
```
