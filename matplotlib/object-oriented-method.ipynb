# Método orientado a objetos

El método orientado a objetos es otra forma de crear graficos pero con mayor personalización.

Un objeto define una figura, la figura es un lienzo donde podemos introducir distintas gráficas (axes), de las cuales cada una tiene sus propios ejes (axis).

![Estructura MOO](../imgs/object-oriented-method.png)

La figura representa un todo, dentro de ella configuramos gráficas con sus propios ejes. 
Este método es un poco más complicado que un plot normal, pero nos brinda una capacidad mucho mayor a la hora de personalizar nuestros graficos.

**Diferencias entre plot y el método orientado a objetos**

| Plot | MOO |
|------|-----|
| Rápido | Mayor personalización |
| Fácil | Más código |
| Una sola figura | Más amigable a multiples diagramas |

## Implementación del método orientado a objetos

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 10)
y = x ** 2
```

**Creamos la figura**

A partir de crear la figura comenzamos a crear los axes, estos reciben los siguientes parametros:

| Pos. Eje X | Pos Eje Y | Tamaño del eje X | Tamaño del eje Y |
|-|-|-|-|

```python
figure = plt.figure() # Lienzo
axes_1 = figure.add_axes([0.1, 0.1, 0.5, 0.9]) # primier axes
axes_1.plot(x, y, 'b:') # grafica
figure.show() 
```

El ejemplo anterior es lo mismo que si pusieramos `plt.plot(x, y, 'b:')`, sin embargo con el MOO podemos agregar más graficos con mayor personalización.

```python
figure = plt.figure() # Lienzo
axes_1 = figure.add_axes([0.1, 0.1, 0.8, 0.9]) # primier axes
axes_2 = figure.add_axes([0.15, 0.5, 0.4, 0.3]) # segundo axes

axes_1.plot(x, y, 'b:') # grafica
axes_2.plot(y, x, 'r--') # grafica

axes_2.set_facecolor('grey')

plt.show() 
```
