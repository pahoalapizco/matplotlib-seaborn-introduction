# Distribuciones

La librería `seaborn` ofrece gaficas y opciones adicionales que permiten trabajar con distribuiciones orientadas a tipos de datos numéricos. 

> :pencil: **Las distribuciones orientadas a tipos de datos numéricos** se refiere a la forma en la que los datos numéricos se distribuyen en una muestra o población.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

## Histogramas: `histplot()`

Parámetros más comunes:
 - **`data`**: DataFrame con los datos.
 - **`x`**: Nombre de la columna del DataFrame a graficar en el eje `x`.
 - **`bins`**: Cantidad de barras a graficar, por default son 15.
 - **`cumulative`**: Representa de forma acumulativa los datos del eje `x`, donde el cantidad de la barra anterior se suma a la siguiente.
 - **`hue`**: Nombre de la columna del DataFrame con los datos por los cuales queremos segmentar la gráfica.
 - **`stat`**: Tipo de representación estadistica:
   - `count`
   - `frequency`
   - `probability`
   - `percent`
   - `density`
 - **`multiple`**: Como queremos representar los datos:
   - `layer`
   - `dodge`
   - `stack`
   - `fill`

Histograma que muestra en el eje `x` las propinas acumuladas (`cumulative`) distribuidas en `20` barras (`bins`) verticales:

```python
sns.histplot(data=tips, x='tip', bins=20, cumulative=True)
plt.show()
```

Histograma que muestra en el eje `x` el total de las propinas segmentadas por genero (`hue`), en esta grafica se representa la densidad (`stat`) de las propinas separando en barras independientes cada segmente(`mutiple`):

```python
sns.histplot(data=tips, x='tip', hue='sex', stat='density', multiple='dodge')
plt.show()
```

## KDE plots: kdeplot()

Gráfico de densidad, **K**ernel **D**ensity **E**stimation, este tipo de gaficas, nos permite visualizar la distribución de las observaciones en un dataset (similar al histograma). El KDE representa los datos utilizando una curva de densidad.

Parametros más comunes:
 - **`data`**: DataFrame con los datos.
 - **`x`**: Nombre de la columna del DataFrame a graficar en el eje `x`.
 - **`cumulative`**: Representa de forma acumulativa los datos del eje `x`, donde el cantidad de la barra anterior se suma a la siguiente.
 - **`hue`**: Nombre de la columna del DataFrame con los datos por los cuales queremos segmentar la gráfica.
 - **`shade`**: Rellena/sombrea el área bajo la curva de densidad, por default el valo es `False` y solo se colorea el périmetro.

KDE que muestra la densidad de las propinas (`x`) segmentadas (`hue`) por genero:

```python
sns.kdeplot(data=tips, x='tip', hue='sex')
plt.show()
```

KDE igual al anterior, pero con su área sombreada (`shade`):

```python
sns.kdeplot(data=tips, x='tip', hue='sex', shade=True)
plt.show()
```

KDE que muestra en el eje `x` las propinas acumuladas (`cumulative`) segmentadas por genero (`hue`):

```python
sns.kdeplot(data=tips, x='tip', hue='sex', shade=True, cumulative=True)
plt.show()
```

## ECDF: ecdfplot()

Esta función representa los datos de forma similar a la propiedad `cumulative` de las graficas anteriores, con la diferencia de que muestra los datos unicos que lo componen mostrando una especie de grafica escalonada.

Parametros más comunes:
 - **`data`**: DataFrame con los datos.
 - **`x`**: Nombre de la columna del DataFrame a graficar en el eje `x`.
 - **`hue`**: Nombre de la columna del DataFrame con los datos por los cuales queremos segmentar la gráfica.

Muestra el acumulado en forma escalonada de las propinas (`x`) segmentadas (`hue`) por fumadores o no fumadores:

```python
sns.ecdfplot(data=tips, x='tip', hue='smoker')
plt.show()
```

## Displot

Con esta función podemos hacer uso de distintas graficas de distribución de datos numéricos. Es una herramienta general que podemos utilizar en sustitución a las funciones individuales de cada grafica.

Parametros más comunes:
 - **`data`**: DataFrame con los datos.
 - **`x`**: Nombre de la columna del DataFrame a graficar en el eje `x`.
 - **`hue`**: Nombre de la columna del DataFrame con los datos por los cuales queremos segmentar la gráfica.
 - **`kind`**: Tipo de grafica, por default muestra un histograma.

Hitograma de las propinas segmentadas por genero:

```python
sns.displot(data=tips, x='tip', hue='sex', multiple='stack')
plt.show()
```

KDE de las propinas segmentadas por genero:

```python
sns.displot(data=tips, x='tip', hue='sex', kind='kde')
plt.show()
```
