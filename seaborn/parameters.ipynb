# Parametros m치s comunes al usar seaborn

Seaborn cuenta con una gran variedad de graficos para distintos propositos, y cada uno de ellos cuenta con una serie de par치metos que nos ayudan a representar la informaci칩n de mejor manera.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Con `load_dataset` cargamos un dataset de exemplo del [repositorio de data](https://github.com/mwaskom/seaborn-data) propio de seaborn, para ello es necesario una conexi칩n a internet.

```python
tips = sns.load_dataset('tips')
tips.head()
```

En este ejemplo se utiliza el tipo de grafica `displot` que nos va a mostrar un histograma con la frecuencia de los datos, donde:
- **data**: tips
- **x**: total_bill

```python
sns.displot(data=tips, x='total_bill')
plt.show()
```

Agregamos el par치metro `y`, donde
- **y**: tips

Seaborn detecta que el grafico de frecuencias no es la mejor opci칩n y lo adapta a una grafica de disitribuci칩n que muestra la relaci칩n entre ambos ejes

```python
sns.displot(data=tips, x='total_bill', y='tip')
plt.show()
```

Con el par치metro `hue` le indicamos como segmentar los datos.
- **hue**: sex

```python
sns.displot(data=tips, x='total_bill', y='tip', hue='sex')
plt.show()
```

`hue` tambi칠n funciona si la grafica es un histograma

```python
sns.displot(data=tips, x='total_bill', hue='sex')
plt.show()
```

`kind` es una propieda con la cual le podemos indicar a la funci칩n `displot` como representar la grafica.

- **kind**: kde, muestra un grafico de densidad 

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde')
plt.show()
```

Con `palette` cambiammos la paleta de colores en el grafico.

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde', palette='Spectral')
plt.show()
```

Con `alpha` cambiamos la opacidad del grafico

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde', palette='Spectral', alpha=0.35)
plt.show()
```
