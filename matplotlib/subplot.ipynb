# Subplot

La función `subplot` del modulo `pyplot` nos permite dibujar mulptiples gráficos en una misma figura, se organizan en forma de tabla.
Sindaxis:
`subplot(row, column, index)` donde:
- `row`: Número de filas.
- `column`: Número de columnas.
- `index`: Indice donde se va a dibujar la grafica.

![Estructura subplot](../imgs/subplot-structure.png)


```python
import matplotlib.pyplot as plt
import numpy as np
```

Generamos los sets de datos aleatorios con los que trabajaremos.

```python
x = np.linspace(0, 8, 11)
y = x ** 2
```

Creamos un subplot de `1` fila y `2` columnas.

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'm--o') # Grafica con línea color morado, marcador en forma de punto u línea doblemente punteada

plt.subplot(1, 2, 2)
plt.hist(x)

plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'm--o') # Plot con línea color morado, marcador en forma de punto y línea doblemente punteada
plt.plot(y, x, 'rD:') # @do Plot con línea color rojo, marcador en forma de diamante y línea punteada

plt.subplot(1, 2, 2)
plt.hist(x)

plt.show()
```

Creamos un subplot de `2` filas y `1` columna.

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'm--o') # Plot con línea color morado, marcador en forma de punto y línea doblemente punteada
plt.plot(y, x, 'rD:') # 2do Plot con línea color rojo, marcador en forma de diamante y línea punteada

plt.subplot(1, 2, 2)
plt.pie(y)

plt.show()
```

Creamos un subplot de `2` filas y `2` columnas.

```python
plt.subplot(2, 2, 1)
plt.plot(x, y, 'm--o') # Plot con línea color morado, marcador en forma de punto y línea doblemente punteada
plt.plot(y, x, 'rD:') # @do Plot con línea color rojo, marcador en forma de diamante y línea punteada

plt.subplot(2, 2, 2)
plt.pie(y)


plt.subplot(2, 2, 3)
plt.triplot(x, y)

plt.subplot(2, 2, 4)
plt.boxplot(y)

plt.show()
```

Creamos un subplot con `2` filas y `2` columnas, la gráfica de la 2 columnas tendrá un span de 2.

```python
plt.subplot(2,2,1)
plt.plot(x,y)

plt.subplot(1,2,2)
plt.hist(x)

plt.subplot(2,2,3)
plt.triplot(y,x)

plt.show()
```
