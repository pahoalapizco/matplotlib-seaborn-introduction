# Categóricos

Seaborn ofrece varios tipos de graficos para representar datos categóricos ó variables de texto. 

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

## Conteo: countplot()

Este tipo de grafica muestra la frecuencia de cada categoría en un conjunto de datos o dataset.

Parámetros más comunes:
 - **`data`**: DataFrame con los datos.
 - **`x`**: Nombre de la columna del DataFrame a graficar en el eje `x`.
 - **`hue`**: Nombre de la columna del DataFrame con los datos por los cuales queremos segmentar la gráfica.

```python
sns.countplot(data=tips, x='day', hue='sex')
plt.show()
```

```python
sns.countplot(data=tips, y='day', hue='sex')
plt.show()
```

## Diagrama de puntos: swarmplot

Esta grafica muestra un punto para cada observación en un dataset.

Parámetros más comunes:
 - **`data`**: DataFrame con los datos.
 - **`x`**: Nombre de la columna del DataFrame a graficar en el eje `x`.
 - **`y`**: Nombre de la columna del DataFrame a graficar en el eje `y`.
 - **`hue`**: Nombre de la columna del DataFrame con los datos por los cuales queremos segmentar la gráfica.
 - **`dodge`**: Separa los datos por segemento, su valor defult es `False`

```python
sns.swarmplot(data=tips, x='day',  y='total_bill', hue='sex')
plt.show()
```

Separamos los puntos por segemente con `dodge`:

```python
sns.swarmplot(data=tips, x='day',  y='total_bill', hue='sex', dodge=True)
plt.show()
```

## Grafica de caja: boxplot()

Muestra la distribución de un conjunto de datos numéricos para cada categoría utilizando cuartiles.

Parámetros más comunes:
 - **`data`**: DataFrame con los datos.
 - **`x`**: Nombre de la columna del DataFrame a graficar en el eje `x`.
 - **`y`**: Nombre de la columna del DataFrame a graficar en el eje `y`.
 - **`hue`**: Nombre de la columna del DataFrame con los datos por los cuales queremos segmentar la gráfica.
 - **`dodge`**: Separa los datos por segemento, su valor defult es `True`.

```python
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

## Graficas de violín: 

Este tipo de graficos muestran la distribución de un conjunto de datos numéricos para cada categoría utilizando un kernel de densidad.

Parámetros más comunes:
 - **`data`**: DataFrame con los datos.
 - **`x`**: Nombre de la columna del DataFrame a graficar en el eje `x`.
 - **`y`**: Nombre de la columna del DataFrame a graficar en el eje `y`.
 - **`hue`**: Nombre de la columna del DataFrame con los datos por los cuales queremos segmentar la gráfica.
 - **`dodge`**: Separa los datos por segemento, su valor defult es `True`.

```python
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex')
plt.show()
```

```python
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', dodge=False)
plt.show()
```

## Graficos categóricos: catplot()

Similar a la función `displot`, `catplot` es de uso general y en ella podemos graficar N tipos de graficos según los parametros que le enviemos.

Parámetros más comunes:
 - **`data`**: DataFrame con los datos.
 - **`x`**: Nombre de la columna del DataFrame a graficar en el eje `x`.
 - **`y`**: Nombre de la columna del DataFrame a graficar en el eje `y`.
 - **`hue`**: Nombre de la columna del DataFrame con los datos por los cuales queremos segmentar la gráfica.
 - **`dodge`**: Separa los datos por segemento, su valor defult es `True`.
 - **`kind`**: Tipo de grafico que deseamos ilustrar.
 - **`col`**: Nombre de la columna por la cual queremos separar las grafias en graficos distintos.

```python
sns.catplot(data= tips, x= 'day', y = 'total_bill', hue= 'sex', dodge=True, kind= 'box', col= 'time')
plt.show()
```
