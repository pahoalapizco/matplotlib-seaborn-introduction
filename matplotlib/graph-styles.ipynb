# Estilos en las graficas

Estilizar las graficas no es solo para hacerlas "bonitas", sino para darle contexto a los datos que estamos representando con cada grafica que dibujemos.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = np.sin(x)
```

## Tamaño, leyendas, etiquetas y títulos

- **Tamaño**: Dimensiones de la figura, se ajusta a aprtiir de la creación de una figura con la propiedad `figsize`
- **leyenda**: Significado de la línea de un grafico, se ajusta a partir de la creación del plot con la propiedad `label` pero se activa con la función `legend()`
- **Etiqueda**: Nombre de los ejes, se asignan con las funciones `set_xlabel` para el eje `X` y `set_ylabel` para el eje `Y`
- **Titulo**: Nombre del grafico, se asigna con la función `set_title('')` de cada axes

**NOTA**: Los títulos, etiquetas y leyendas acepta notación matemática creada con `TeX` utilizando signos de pesos al inicio y fin del texto. 

```python
figure, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5)) # fisize cambia el tamaño de la figura completa
ax1.plot(x, y, 'r:<', label='$sin(x)$')
ax1.legend() # Activa la leyenda (significado de la linea en el grafico)
# Titulos
ax1.set_title('Relacion X - Y') # Titulo del grafico
ax1.set_xlabel('X') # Etiqueta del eje X
ax1.set_ylabel('Y') # Etiqueta del eje Y

ax2.plot(y, x, 'g--x', label='sin(x)')
ax2.legend() # Activa la leyenda (significado de la linea en el grafico)
# Titulos
ax2.set_title('Relacion Y - X')
ax2.set_xlabel('Y') # Etiqueta del eje X
ax2.set_ylabel('X') # Etiqueta del eje Y
```

## Colores y estilos

### Estilos de graficas

`pyplot` maneja varios estilos de graficos.

Para ver aquellos disponibles en la libreria:

```python
print(plt.style.available)
```

Para utilizar cualquiera de los estilos disponibles, se le indica con la función `style.use()` enviando como parametro el nombre del estilo.

```python
plt.style.use('dark_background')
```

```python
fig, axes = plt.subplots()
```

### Colores y estilos

#### Matlab

```python
figure, ax = plt.subplots(figsize = (6,6))
ax.plot(x,x+1, 'r--')
ax.plot(x,x+2, 'bo-')
ax.plot(x,x+3, 'g.:')
ax.plot(x,x+4, 'purple')
```

#### Pyplot

**Colores**

```python
figure, ax = plt.subplots(figsize = (6,6))
ax.plot(x,x+1,color = '#D426C8') # Color RGB
ax.plot(x,x+2,color = '#66FF89')

ax.plot(x,x+3,color = 'blue') # Color común
ax.plot(x,x+4, color = 'white')
```

**Tansparencia `(alpha)` y grosor `(linewidth)`**

- **alpha**: El nivel de transparencia esta entre 0 y 1, entre más nos acerquemos al 0 mayor será la transparencia de la linea. 

```python
figure, ax = plt.subplots(figsize = (6,6))
ax.plot(x,x+1,color = '#D426C8', alpha=0.2, linewidth=8) # Color RGB
ax.plot(x,x+2,color = '#66FF89', linewidth=1.2)

ax.plot(x,x+3,color = 'blue', linewidth=6.8) # Color común
ax.plot(x,x+4, color = 'white', alpha=0.4)
```

**Estilo `(linestyle)` y marcadores `(marker)`**

```python
fig, ax = plt.subplots(figsize = (6,6))
ax.plot(x,x+2, color = '#ffbb11',linestyle = ':', marker = '+')
ax.plot(x,x+3,color = 'blue', linestyle = 'dashed', marker = 'o')
ax.plot(x,x+4, color = 'purple',linestyle = '--', marker = 'P')
```

**Tamaño de marcadores `(markersize)` y color `(markerfacecolor)`**

```python
fig, ax = plt.subplots(figsize = (6,6))
ax.plot(x,x+1,color = 'blue', linestyle = 'dashed', marker = '8', markersize = 10, markerfacecolor= "#37D842")

ax.plot(x,x+2, color = 'white',linestyle = '--', marker = 'P', markerfacecolor="#FF0000")
```

```python
fig, ax = plt.subplots(figsize = (6,6))
ax.plot(x,x+3,color = 'blue', linestyle = 'dashed', marker = '8', markersize = 10, markerfacecolor= "#37D842")

ax.plot(x,x+4, color = 'white',linestyle = '--', marker = 'P', markerfacecolor="#FF0000")
```
