# Más variedad de graficos

Existen distintos tipos de graficos ademas de las líneas y barras. 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1, 50, 100)
data
```

## Histograma
Se utiliza para representar la distribución de frecuencias de algunos puntos de datos de una variable.

```python
plt.hist(data)
plt.show()
```

**Cambiar número de barras `(bins)`**
- `bins`: Número de barras a graficar de igual tamaño, su valor por default es 10

Muestra la frecuencia de los datos por segmento.

```python
plt.hist(data, bins=50)
plt.show()
```

**Cambiar stilo de las barras**
- `histtype`: Por default toma el valor 'bar' que rellena el área de las barras.

```python
fig, (ax_r1, ax_r2) = plt.subplots(2, 2)

# primer row
ax_r1[0].set_title('bar')
ax_r1[0].hist(data, histtype='bar')

ax_r1[1].set_title('barstacked')
ax_r1[1].hist(data, histtype='barstacked')

# segundo row
ax_r2[0].set_title('step')
ax_r2[0].hist(data, histtype='step')

ax_r2[1].set_title('stepfilled')
ax_r2[1].hist(data, histtype='stepfilled')

fig.tight_layout() 
```

## Boxplot
Representa gráficamente una serie de datos numéricos a través de sus cuartiles. De esta manera, se muestran a simple vista la mediana y los cuartiles de los datos,​ y también pueden representarse sus valores atípicos.

- **cuartiles**: Medidas de localización, su función es informar del valor de la variable que ocupará la posición (en tanto por cien) que nos interese respecto de todo el conjunto de variables.

```python
plt.boxplot(data)
plt.show()
```

**Boxplot horizontal**

```python
plt.boxplot(data, vert=False)
plt.show()
```

**Rellenar el rango inter cuartil (`patch_artist`) y enfocar la mediana (`notch`)**

```python
plt.boxplot(data, vert=False, patch_artist=True, notch=True)
plt.show()
```

**Valores atipicos (outliers)**

```python
data = np.append(data, 80)

plt.boxplot(data, vert=False, patch_artist=True, notch=True)
plt.show()
```

## Scatter
Utilizado para representar la relación entre dos variables. <br>
En español se conoce como "_diagrama de dispersión_".

```python
N = 50 
x = np.random.rand(N) # Coordenadas en el eje x
y = np.random.rand(N) # Coordenadas en el eje y
area = (30 * np.random.rand(N)) ** 2 # Area de cada punto en el plano
```

```python
plt.scatter(x, y)
plt.show
```

**Área**

```python
plt.scatter(x, y, s=area)
plt.show
```

**Colores y markers**

```python
colors = np.random.rand(N)
```

```python
fig, (ax_r1, ax_r2) = plt.subplots(2,2)

ax_r1[0].scatter(x, y, s=area, c=colors, marker='X')
ax_r1[1].scatter(x, y, s=area, c=colors, marker='<')

ax_r2[0].scatter(x, y, s=area, c=colors, marker='p')
ax_r2[1].scatter(x, y, s=area, c=colors, marker='3')
```

**Transparencia**

Ideal para ver los valores superpuestos

```python
fig, (ax_r1, ax_r2) = plt.subplots(2,2)

ax_r1[0].scatter(x, y, s=area, c=colors, marker='X', alpha=0.5)
ax_r1[1].scatter(x, y, s=area, c=colors, marker='<', alpha=0.5)

ax_r2[0].scatter(x, y, s=area, c=colors, marker='p', alpha=0.5)
ax_r2[1].scatter(x, y, s=area, c=colors, marker='3', alpha=0.5)
```
