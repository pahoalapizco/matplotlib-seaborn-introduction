# Joinplot y Pairplot

Ambos gráficos pertenecen al grupo de `relational`, con la diferencia de que con ellos podemos visualizar más de un gráfico a la vez.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

## Joinplot

Este gráfico combina dos gráficos diferentes: un scatter y un histograma. Donde, el scatter plot muestra la correlación entre dos variables numéricas, mientras que el histograma muestra la distribución de cada variable por separado. Útil para visualizar distribución y correlación entre dos variables numéricas al mismo tiempo.

Parámetros más comunes:
 - **`data`**: DataFrame con los datos.
 - **`x`**: Nombre de la columna del DataFrame de donde tomará el valor para la coordenada `x`.
 - **`y`**: Nombre de la columna del DataFrame de donde tomará el valor para la coordenada `y`.
 - **`hue`**: Nombre de la columna del DataFrame con los datos por los cuales queremos segmentar la gráfica.
 - **`kind`**: Especificar el tipo de gráfico a ilustrar.
 - **`marginal_ticks`**: Muestra una pequeña tabla para la gráficas exteriores (histogramas).
 - **`marginal_kws`**: Agregar distintos parametros de configuración a las gráficas exteriores (histogramas).

Correlación y distribución entre `total_bill` y `tip`:

```python
sns.jointplot(data=tips, x='total_bill', y='tip');
```

Segmentado por genero:

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex');
```

Le agregamos información adicional a los histogramas y los configuramos para que muestren informacián más detallada.

```python
config_hist = dict(bins=20, fill=True, multiple='layer')
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='hist', marginal_ticks=True, 
              marginal_kws=dict(bins=25, fill=False, multiple='dodge'));
```

# Pairplot

Este gráfico muestra la relación entre todas las variables numéricas de un dataset. Crea una matriz de scatterplots e histogramas, donde cada scatter muestra la correlación entre dos variables numéricas y los histogramas representan la distribución de cada variable.
Útiles para visualizar la relación entre todas las variables numéricas.

Parámetros más comunes:
 - **`data`**: DataFrame con los datos.
 - **`hue`**: Nombre de la columna del DataFrame con los datos por los cuales queremos segmentar la gráfica.
 - **`kind`**: Especificar el tipo de gráfico a ilustrar.
 - **`corner`**: Elimina de la visualización las gráficas repetidas.

Visualización de todas las variables numéricas segmentadas por día:

```python
sns.pairplot(data=tips, hue='day');
```

Eliminamos las graficas repetidas:

```python
sns.pairplot(data=tips, corner=True);
```

Cambiamos el tipo de gráfico:

```python
sns.pairplot(data=tips, hue='sex', corner=True, kind='hist');
```
